### 4.3 Loading and Reading CSV files

```python
# add Matplotlib inline magic command
%matplotlib inline

# dependencies and setup
import matplotlib.pyplot as plt
import pandas as pd
import os


# file to Load (Remember to change these)
city_data_to_load = os.path.join("Resources", "city_data.csv")
ride_data_to_load = os.path.join("Resources", "ride_data.csv")

# read the city and ride data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)

# see top 5 in ride data, test
ride_data_df.head()
```

```python
# see top 5 in city data, test

city_data_df.head()
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
# get the total rides for each city type

total_rides_city_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides_city_type
```

```python
# get the total drivers for each city type

total_drivers_city_type = pyber_data_df.groupby(["type"]).count()["driver_count"]
total_drivers_city_type
```

```python
#  get the total amount of fares for each city type

total_fares_city_type = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares_city_type
```

```python
#  get the average fare per ride for each city type

average_ride_fare_city_type = total_fares_city_type / total_rides_city_type
average_ride_fare_city_type
```

```python
# get the average fare per driver for each city type

average_fare_driver_city_type = total_fares_city_type / total_drivers_city_type
average_fare_driver_city_type
```

```python
#  create a PyBer summary DataFrame

pyber_summary_df = pd.DataFrame()

pyber_summary_df["Total Rides"] = total_rides_city_type
pyber_summary_df["Total Drivers"] = total_drivers_city_type
pyber_summary_df["Total Fares"] = total_fares_city_type
pyber_summary_df["Average Fare Per Ride"] = average_ride_fare_city_type
pyber_summary_df["Average Fare Per Driver"] = average_fare_driver_city_type

pyber_summary_df
```

```python
#  cleaning up the dataframe, delete the index name

pyber_summary_df.index.name = None
```

```python
pyber_summary_df.dtypes
```

```python
pyber_summary_df["Average Fare Per Driver"].map("${:.2f}".format)
```

```python
#  format the columns

pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,}".format)
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,}".format)
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare Per Ride"] = pyber_summary_df["Average Fare Per Ride"].map("${:.2f}".format)
pyber_summary_df["Average Fare Per Driver"] = pyber_summary_df["Average Fare Per Driver"].map("${:.2f}".format)

pyber_summary_df


```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# read the merged dataframe

pyber_data_df.head()
```

```python
# using groupby() to create a new dataframe showing the sum of the fares 
#  for each date where the indices are the city type and date

fare_city_date_df = pyber_data_df.groupby(["type", "date"]).sum().fare.to_frame()
fare_city_date_df
```

```python
# reset the index on the DataFrame you created in #1, this is needed to use the 'pivot()' function
# df = df.reset_index()

fare_city_date_df = fare_city_date_df.reset_index()
fare_city_date_df
```

```python
# create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date

fare_city_date_df = fare_city_date_df.pivot(index="date", columns="type", values="fare")
fare_city_date_df
```

```python
# create a new dataframe from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'

fare_date_df = fare_city_date_df.loc['2019-01-01':'2019-04-29']
fare_date_df
```

```python
# set the "date" index to datetime datatype. this is necessary to use the resample() method in Step 8
# df.index = pd.to_datetime(df.index)

type(fare_date_df.index)
fare_date_df.index = pd.to_datetime(fare_date_df.index)

fare_date_df
```

```python
# check that the datatype for the index is datetime using df.info()

fare_date_df.info()
```

```python
# create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week

fare_date_df = fare_date_df.resample('W').sum()
fare_date_df
```

```python
# using the object-oriented interface method, plot the resample DataFrame using the df.plot() function

# Import the style from Matplotlib
from matplotlib import style

# Use the graph style fivethirtyeight
style.use('fivethirtyeight')

fare_date_df.plot(figsize=(18, 5))
plt.ylabel('Fare (USD)')
plt.title('Total Fare by City Type')
plt.savefig('analysis/pyber_summary.png');
```
